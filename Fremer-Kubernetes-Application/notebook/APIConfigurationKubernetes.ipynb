```python
import subprocess

deployment_name = "concert-ticket-deployment"
namespace = "default"

try:
    output = subprocess.check_output(
        ["kubectl", "get", "deployment", deployment_name, "-n", namespace, "-o=jsonpath={.status.replicas}"],
        text=True,
        stderr=subprocess.STDOUT
    ).strip()
    replika = int(output) if output else 0
    print("Current replicas:", replika)
except subprocess.CalledProcessError as e:
    print("❌ Error running kubectl:", e.output.strip())
    replika = 0
```

```python
def checkreplicas():
  deployment_name = "concert-ticket-deployment"
  namespace = "default"

  # Get replica count from kubectl
  output = subprocess.check_output(
      ["kubectl", "get", "deployment", deployment_name, "-n", namespace, "-o=jsonpath={.status.replicas}"],
      text=True
  )

  replika = int(output) if output else 0
  return replika
```

```python
MAX_PODS = 5
MIN_PODS = 1

def upscale():
    current = checkreplicas()
    if current < MAX_PODS:
        new_replika = current + 1
        subprocess.run(
            ["kubectl", "scale", "deployment", deployment_name, f"--replicas={new_replika}", "-n", namespace],
            check=True
        )
        print(f"✅ Success Upscale: {deployment_name} now has {checkreplicas()} replicas")
    else:
        print("⚠️ Max Pods Reached")

def downscale():
    current = checkreplicas()
    if current > MIN_PODS:
        new_replika = current - 1
        subprocess.run(
            ["kubectl", "scale", "deployment", deployment_name, f"--replicas={new_replika}", "-n", namespace],
            check=True
        )
        print(f"✅ Success Downscale: {deployment_name} now has {checkreplicas()} replicas")
    else:
        print("⚠️ Min Pods Reached")
```

```python
import requests
import pandas as pd
import matplotlib.pyplot as plt
from datetime import datetime, timezone
import re

# --- Prometheus connection ---
PROM_URL = "http://34.72.255.30:80"   # your Prometheus LoadBalancer IP

# --- Time range: last 1 hour ---
end = int(datetime.now(timezone.utc).timestamp())
start = end - 3600
step = "30s"

# --- Helper to safely query Prometheus ---
def prom_query_range(query):
    try:
        resp = requests.get(
            f"{PROM_URL}/api/v1/query_range",
            params={"query": query, "start": start, "end": end, "step": step},
            timeout=10
        )
        resp.raise_for_status()
        return resp.json()
    except Exception as e:
        print(f"❌ Error fetching from Prometheus: {e}")
        return {"data": {"result": []}}

# --- Convert Prometheus range data to DataFrame ---
def prom_range_to_df(prom_response, metric_name):
    results = prom_response.get("data", {}).get("result", [])
    rows = []
    for r in results:
        pod = r["metric"].get("pod", "unknown")
        for v in r.get("values", []):
            ts = int(v[0])
            val = float(v[1]) if v[1] != "NaN" else 0
            rows.append({
                "timestamp": ts,
                "pod": pod,
                metric_name: val
            })
    if not rows:
        return pd.DataFrame(columns=["timestamp", "pod", metric_name])
    return pd.DataFrame(rows)

# --- Auto-detect pod prefix from Prometheus ---
def detect_pod_prefix():
    try:
        resp = requests.get(f"{PROM_URL}/api/v1/query", params={
            "query": 'count(container_cpu_usage_seconds_total{namespace="default"}) by (pod)'
        }).json()
        pods = [r["metric"]["pod"] for r in resp["data"]["result"] if "pod" in r["metric"]]
        # find something like 'concert-ticket-deployment'
        matches = [re.match(r"([a-z0-9-]+)-[a-z0-9]{5,10}$", p) for p in pods]
        bases = [m.group(1) for m in matches if m]
        if bases:
            prefix = max(set(bases), key=bases.count)
            print(f"🔍 Detected pod prefix: {prefix}")
            return prefix
        print("⚠️ Could not detect pod prefix, defaulting to 'deployment-1'")
        return "deployment-1"
    except Exception as e:
        print(f"❌ Error detecting pods: {e}")
        return "deployment-1"

# --- Detect your deployment's pod name pattern ---
pod_prefix = detect_pod_prefix()

# --- Build queries dynamically ---
query_cpu = f'''
sum by (pod) (
  rate(container_cpu_usage_seconds_total{{
    namespace="default",
    container!="",
    pod=~"{pod_prefix}.*"
  }}[5m])
)
'''

query_mem = f'''
sum by (pod) (
  container_memory_usage_bytes{{
    namespace="default",
    container!="",
    pod=~"{pod_prefix}.*"
  }}
)
'''

# --- Fetch metrics ---
resp_cpu = prom_query_range(query_cpu)
df_cpu = prom_range_to_df(resp_cpu, "cpu_usage")

resp_mem = prom_query_range(query_mem)
df_mem = prom_range_to_df(resp_mem, "memory")

if not df_mem.empty:
    df_mem["memory"] = df_mem["memory"] / 1024 / 1024  # bytes → MB

# --- Merge & aggregate ---
df = pd.merge(df_cpu, df_mem, on=["timestamp", "pod"], how="outer")
if not df.empty:
    df["date"] = pd.to_datetime(df["timestamp"], unit="s", utc=True).dt.tz_convert("Asia/Jakarta")

    df_final = df.groupby("date", as_index=False).agg({
        "cpu_usage": "sum",
        "memory": "sum"
    }).sort_values("date")

    df_final.to_csv(f"{pod_prefix}_usage_1h.csv", index=False)

    print("✅ Data 1 jam terakhir berhasil diambil dari Prometheus.")
    print(df_final.head())

    fig, ax1 = plt.subplots(figsize=(10,4))
    ax1.plot(df_final["date"], df_final["cpu_usage"], label="CPU Usage (cores)", color="tab:blue")
    ax2 = ax1.twinx()
    ax2.plot(df_final["date"], df_final["memory"], label="Memory (MB)", color="tab:red")

    ax1.set_xlabel("Time (WIB)")
    ax1.set_ylabel("CPU (cores)", color="tab:blue")
    ax2.set_ylabel("Memory (MB)", color="tab:red")
    plt.title(f"{pod_prefix}: CPU & Memory Usage (Last 1 Hour)")
    plt.tight_layout()
    plt.show()
else:
    print("⚠️ No data found for the last hour.")
```

```python
#GET DATA 3 HARI CLEAN
#=========================================
import requests
import pandas as pd
import numpy as np
from datetime import datetime, timezone
import re

# ==========================
# 1. Prometheus Config
# ==========================
PROM_URL = "http://34.72.255.30:80"  # ganti dengan IP Prometheus

# --- Waktu sekarang dan 3 hari ke belakang ---
end = int(datetime.now(timezone.utc).timestamp())
start = end - (10 * 24 * 3600)  # 10 hari ke belakang
step = "600s"  # interval setiap 10 menit

# ==========================
# 2. Query Prometheus
# ==========================
def prom_query_range(query):
    try:
        resp = requests.get(
            f"{PROM_URL}/api/v1/query_range",
            params={"query": query, "start": start, "end": end, "step": step},
            timeout=30
        )
        resp.raise_for_status()
        return resp.json()
    except Exception as e:
        print(f"❌ Error fetching from Prometheus: {e}")
        return {"data": {"result": []}}

def detect_pod_prefix():
    try:
        resp = requests.get(f"{PROM_URL}/api/v1/query", params={
            "query": 'count(container_cpu_usage_seconds_total{namespace="default"}) by (pod)'
        }).json()
        pods = [r["metric"]["pod"] for r in resp["data"]["result"] if "pod" in r["metric"]]
        matches = [re.match(r"([a-z0-9-]+)-[a-z0-9]{5,10}$", p) for p in pods]
        bases = [m.group(1) for m in matches if m]
        if bases:
            prefix = max(set(bases), key=bases.count)
            print(f"🔍 Detected pod prefix: {prefix}")
            return prefix
        print("⚠️ Could not detect pod prefix, defaulting to 'deployment-1'")
        return "deployment-1"
    except Exception as e:
        print(f"❌ Error detecting pods: {e}")
        return "deployment-1"

def prom_range_to_df(prom_response):
    results = prom_response.get("data", {}).get("result", [])
    rows = []
    for r in results:
        pod = r["metric"].get("pod", "unknown")
        for v in r.get("values", []):
            ts = int(v[0])
            val = float(v[1]) if v[1] != "NaN" else 0
            rows.append({"timestamp": ts, "pod": pod, "data": val})
    if not rows:
        return pd.DataFrame(columns=["timestamp", "pod", "data"])
    return pd.DataFrame(rows)

# ==========================
# 3. Ambil Data dari Prometheus
# ==========================
pod_prefix = detect_pod_prefix()

query_cpu = f'''
sum by (pod) (
  rate(container_cpu_usage_seconds_total{{
    namespace="default",
    container!="",
    pod=~"{pod_prefix}.*"
  }}[5m])
)
'''

resp_cpu = prom_query_range(query_cpu)
df_cpu = prom_range_to_df(resp_cpu)

# ==========================
# 4. Preprocessing Data
# ==========================
if df_cpu.empty:
    raise ValueError("⚠️ Tidak ada data CPU yang berhasil diambil dari Prometheus.")

# konversi timestamp
df_cpu["date"] = pd.to_datetime(df_cpu["timestamp"], unit="s", utc=True).dt.tz_convert("Asia/Jakarta")
df_cpu["cols"] = "instance_" + df_cpu["pod"].astype(str)
df_cpu["date"] = df_cpu["date"].dt.floor("10min")
df_cpu = df_cpu.groupby(["date"], as_index=False)["data"].mean()

# pastikan kolom data bersih
df_cpu['data'] = df_cpu['data'].astype(str).str.replace('%', '', regex=False)
df_cpu['data'] = df_cpu['data'].str.replace(',', '.', regex=False)
df_cpu['data'] = pd.to_numeric(df_cpu['data'], errors='coerce')
df_cpu['data'] = df_cpu['data'].fillna(method='ffill')

# deteksi interval dominan
interval = df_cpu['date'].diff().dropna().mode()[0]
print(f"🕒 Interval dominan terdeteksi: {interval}")

# buat rentang waktu penuh
full_range = pd.date_range(start=df_cpu['date'].min(), end=df_cpu['date'].max(), freq=interval)
df_cpu = df_cpu.set_index('date').reindex(full_range)
df_cpu.index.name = 'date'

# interpolasi missing value
df_cpu['data'] = df_cpu['data'].interpolate(method='time')

# normalisasi (kalau datanya kecil)
if df_cpu['data'].max() < 1:
    print("📈 Nilai data kecil (<1), dikonversi ke skala persentase.")
    df_cpu['data'] = df_cpu['data'] * 100

# buang outlier ekstrem
df_cpu['data'] = df_cpu['data'].clip(
    lower=df_cpu['data'].quantile(0.001),
    upper=df_cpu['data'].quantile(0.999)
)

# tambahkan kolom 'cols'
df_cpu['cols'] = 'cpu_usage'

# ==========================
# 5. Simpan hasil
# ==========================
output_path = "/content/cpu_10_hari.csv"
df_cpu.to_csv(output_path, index=True)

print("✅ File bersih 10 hari ke belakang berhasil dibuat:", output_path)
print("Jumlah data:", len(df_cpu))
print("Rentang waktu:", df_cpu.index.min(), "→", df_cpu.index.max())
print("\nStatistik kolom data:")
print(df_cpu['data'].describe())
print("\nCek interval unik:")
print(df_cpu.index.to_series().diff().value_counts().head())
```

```python
import pandas as pd
import numpy as np
from datetime import datetime, timedelta

# Konfigurasi dataset
start_date = datetime(2025, 11, 1, 0, 0, 0)
end_date = start_date + timedelta(days=10)
interval = timedelta(minutes=10)

timestamps = pd.date_range(start=start_date, end=end_date, freq=interval)
np.random.seed(42)

cpu_usages = []

for t in timestamps:
    day = (t - start_date).days + 1
    hour = t.hour + t.minute/60.0
    weekday = t.weekday()  # 0=Senin ... 6=Minggu

    # Tentukan fase berdasarkan kalender (tanpa kolom tambahan)
    if (weekday in [5, 6]) and day <= 2:  # 1–2 Nov weekend awal
        min_cpu, max_cpu = 2.0, 5.0
    elif 3 <= day <= 7:  # weekday
        min_cpu, max_cpu = 1.0, 4.0
    elif 8 <= day <= 9:  # weekend menjelang event
        min_cpu, max_cpu = 3.0, 6.0
    else:  # hari 10 = event utama
        min_cpu, max_cpu = 4.0, 7.0

    # Pola harian halus (siang tinggi, malam rendah)
    base_cycle = (np.sin((hour - 8) / 24 * 2 * np.pi) + 1) / 2
    base = min_cpu + base_cycle * (max_cpu - min_cpu)
    noise = np.random.normal(0, 0.1)
    cpu_val = max(0, base + noise)

    # Spike tambahan saat jam sibuk di hari event
    if day == 10 and (11 <= hour <= 13 or 18 <= hour <= 20):
        cpu_val += np.random.uniform(0.8, 1.5)

    cpu_usages.append(round(min(cpu_val, 7.0), 6))

# Buat DataFrame (3 kolom saja)
df_final = pd.DataFrame({
    "date": pd.to_datetime(timestamps).tz_localize("Asia/Jakarta"),
    "data": cpu_usages,
    "cols": ["cpu_usage"] * len(cpu_usages)
})

# Simpan ke CSV
df_final.to_csv("cpu_usage_10hari_calendar_realistic_3cols.csv", index=False)

print("✅ File berhasil dibuat: cpu_usage_10hari_calendar_realistic_3cols.csv")
print(df_final.head(12))
print(df_final.tail(12))
```

```python
# ==========================================================
# 6. Live Prediction + Autoscale Loop (dengan Min-Max Pods)
# ==========================================================
import sys
import torch
import time
import math
import subprocess
import numpy as np
import pandas as pd
import requests
from datetime import datetime, timezone

# ======== Sesuaikan path ke folder Fremer ========
sys.path.append("/content/Fremer-Masterr/Fremer-master")

# Setelah itu baru import
from ts_benchmark.baselines.fremer.fremer_model import FremerModel

# ----- konfigurasi model & sistem -----
MODEL_PATH = "/content/checkpoints/Fremer_cpu_usage_10hari_realistic_3cols_sl720_pl144/Fremer_best.pth"
SEQ_LEN = 720   # panjang input model
PRED_LEN = 144  # jumlah step prediksi
INTERVAL_LOOP = 30  # interval loop (30 detik)

PROM_URL = "http://34.72.255.30:80/api/v1/query_range"
query_cpu = 'avg(rate(container_cpu_usage_seconds_total[5m])) * 100'

# konfigurasi autoscale
DEPLOYMENT_NAME = "concert-ticket-deployment"
NAMESPACE = "default"
MIN_PODS = 1
MAX_PODS = 10
TARGET_CPU = 50  # CPU ideal per pod (%)

# ==========================================================
# Load Fremer Model
# ==========================================================
torch.serialization.add_safe_globals([FremerModel])

print("🧠 Loading Fremer model...")
model = torch.load(MODEL_PATH, map_location="cpu", weights_only=False)
model.eval()
print("✅ Model siap digunakan untuk prediksi real-time\n")

# ==========================================================
# Fungsi tambahan Prometheus
# ==========================================================
def prom_query_range(query):
    """Ambil data dari Prometheus API."""
    end = int(datetime.now(timezone.utc).timestamp())
    start = end - (5 * 24 * 3600)  # 5 hari ke belakang
    step = "10m"

    params = {"query": query, "start": start, "end": end, "step": step}
    resp = requests.get(PROM_URL, params=params)
    resp.raise_for_status()
    return resp.json()

def prom_range_to_df(resp):
    """Konversi hasil Prometheus ke DataFrame."""
    if "data" not in resp or "result" not in resp["data"] or len(resp["data"]["result"]) == 0:
        return pd.DataFrame(columns=["timestamp", "data"])
    values = resp["data"]["result"][0]["values"]
    df = pd.DataFrame(values, columns=["timestamp", "data"])
    df["timestamp"] = df["timestamp"].astype(float)
    df["data"] = df["data"].astype(float)
    return df

# ==========================================================
# Fungsi Kubernetes Helper
# ==========================================================
def checkreplicas():
    """Ambil jumlah pods saat ini dari kubectl."""
    try:
        output = subprocess.check_output(
            ["kubectl", "get", "deployment", DEPLOYMENT_NAME, "-n", NAMESPACE, "-o=jsonpath={.status.replicas}"],
            text=True,
            stderr=subprocess.STDOUT
        ).strip()
        return int(output) if output else 0
    except subprocess.CalledProcessError as e:
        print("❌ Error running kubectl:", e.output.strip())
        return 0

def setreplicas(count):
    """Set jumlah pod dengan kubectl scale."""
    try:
        subprocess.run(
            ["kubectl", "scale", "deployment", DEPLOYMENT_NAME, f"--replicas={count}", "-n", NAMESPACE],
            check=True,
            text=True
        )
        print(f"✅ Deployment diubah menjadi {count} pods.")
    except subprocess.CalledProcessError as e:
        print("⚠️ Gagal mengubah pods:", e)

# ==========================================================
# Fungsi Prediksi & Autoscale
# ==========================================================
def predict_cpu(df_cpu):
    data = df_cpu["data"].values.astype(np.float32)
    if len(data) < SEQ_LEN:
        raise ValueError("Data CPU tidak cukup untuk prediksi.")

    data_norm = (data - np.mean(data)) / np.std(data)
    x_input = torch.tensor(data_norm[-SEQ_LEN:]).unsqueeze(0).unsqueeze(-1)

    with torch.no_grad():
        pred = model(x_input).cpu().numpy().flatten()

    pred_real = pred * np.std(data) + np.mean(data)
    return pred_real

def autoscale_logic(pred_cpu):
    """Menentukan scaling berdasarkan hasil prediksi CPU."""
    avg_pred = np.mean(pred_cpu)
    current_pods = checkreplicas()

    # Hitung jumlah pod ideal berdasarkan target CPU
    desired_pods = math.ceil(avg_pred / TARGET_CPU)
    desired_pods = max(MIN_PODS, min(desired_pods, MAX_PODS))

    if desired_pods > current_pods:
        print(f"🚀 CPU tinggi ({avg_pred:.2f}%), SCALE UP dari {current_pods} → {desired_pods}")
        setreplicas(desired_pods)
    elif desired_pods < current_pods:
        print(f"🧊 CPU rendah ({avg_pred:.2f}%), SCALE DOWN dari {current_pods} → {desired_pods}")
        setreplicas(desired_pods)
    else:
        print(f"⚖️ CPU stabil ({avg_pred:.2f}%), tetap di {current_pods} pods.")

# ==========================================================
# Loop Utama
# ==========================================================
while True:
    try:
        print("\n==========================================")
        print("⏰", datetime.now().strftime("%Y-%m-%d %H:%M:%S"), " - Memperbarui data Prometheus...")

        resp_cpu = prom_query_range(query_cpu)
        df_cpu_live = prom_range_to_df(resp_cpu)

        if df_cpu_live.empty:
            print("⚠️ Tidak ada data baru. Coba lagi nanti.")
            time.sleep(INTERVAL_LOOP)
            continue

        # preprocessing
        df_cpu_live["date"] = pd.to_datetime(df_cpu_live["timestamp"], unit="s", utc=True).dt.tz_convert("Asia/Jakarta")
        df_cpu_live["date"] = df_cpu_live["date"].dt.floor("10min")
        df_cpu_live = df_cpu_live.groupby(["date"], as_index=False)["data"].mean().set_index("date")
        df_cpu_live['data'] = df_cpu_live['data'].fillna(method='ffill').clip(lower=0)

        # Ambil nilai CPU real saat ini (nilai terakhir)
        real_cpu_now = df_cpu_live["data"].iloc[-1]
        print(f"📊 CPU real saat ini: {real_cpu_now:.2f}%")

        # === prediksi ===
        pred_real = predict_cpu(df_cpu_live)

        # === hasil prediksi ===
        last_ts = df_cpu_live.index[-1]
        future_timestamps = pd.date_range(start=last_ts, periods=PRED_LEN+1, freq="10min")[1:]
        df_pred = pd.DataFrame({"timestamp": future_timestamps, "predicted_cpu": pred_real})

        print(df_pred.head(5))
        print("...")
        print(df_pred.tail(3))

        # === eksekusi autoscale ===
        autoscale_logic(pred_real)

    except Exception as e:
        print(f"❌ Error dalam loop: {e}")

    print("\n⏳ Menunggu 30 detik sebelum prediksi berikutnya...\n")
    time.sleep(INTERVAL_LOOP)
```

```python
import sys
import torch
import time
import math
import subprocess
import numpy as np
import pandas as pd
from datetime import datetime, timedelta

# ======== Path ke Fremer & Model ========
sys.path.append("/content/Fremer-Masterr/Fremer-master")
from ts_benchmark.baselines.fremer.fremer_model import FremerModel

MODEL_PATH = "/content/checkpoints/Fremer_cpu_usage_1bulan_sl720_pl144/Fremer_best.pth"
CSV_PATH = "/content/drive/MyDrive/dataset/forecasting/forecasting/cpu_usage_1bulan.csv"  # <= CSV buatan kamu

SEQ_LEN = 720   # panjang input
PRED_LEN = 144  # jumlah prediksi
INTERVAL_LOOP = 30  # detik per iterasi

# konfigurasi autoscale
DEPLOYMENT_NAME = "concert-ticket-deployment"
NAMESPACE = "default"
MIN_PODS = 1
MAX_PODS = 10
TARGET_CPU = 50  # target CPU (%)

# ==========================================================
# Load Model
# ==========================================================
torch.serialization.add_safe_globals([FremerModel])

print("🧠 Loading Fremer model...")
model = torch.load(MODEL_PATH, map_location="cpu", weights_only=False)
model.eval()
print("✅ Model siap digunakan untuk prediksi dari CSV\n")

# ==========================================================
# Helper Kubernetes
# ==========================================================
def checkreplicas():
    try:
        output = subprocess.check_output(
            ["kubectl", "get", "deployment", DEPLOYMENT_NAME, "-n", NAMESPACE, "-o=jsonpath={.status.replicas}"],
            text=True,
            stderr=subprocess.STDOUT
        ).strip()
        return int(output) if output else 0
    except subprocess.CalledProcessError as e:
        print("❌ Error kubectl:", e.output.strip())
        return 0

def setreplicas(count):
    try:
        subprocess.run(
            ["kubectl", "scale", "deployment", DEPLOYMENT_NAME, f"--replicas={count}", "-n", NAMESPACE],
            check=True,
            text=True
        )
        print(f"✅ Deployment diubah menjadi {count} pods.")
    except subprocess.CalledProcessError as e:
        print("⚠️ Gagal mengubah pods:", e)

# ==========================================================
# Fungsi Prediksi & Autoscale
# ==========================================================
def predict_cpu(df_cpu):
    data = df_cpu["data"].values.astype(np.float32)
    if len(data) < SEQ_LEN:
        raise ValueError("Data CPU tidak cukup untuk prediksi.")

    data_norm = (data - np.mean(data)) / np.std(data)
    x_input = torch.tensor(data_norm[-SEQ_LEN:]).unsqueeze(0).unsqueeze(-1)

    with torch.no_grad():
        pred = model(x_input).cpu().numpy().flatten()

    pred_real = pred * np.std(data) + np.mean(data)
    return pred_real

def autoscale_logic(real_cpu, pred_cpu):
    avg_pred = np.mean(pred_cpu)
    current_pods = checkreplicas()
    desired_pods = math.ceil(avg_pred / TARGET_CPU)
    desired_pods = max(MIN_PODS, min(desired_pods, MAX_PODS))

    print(f"📊 CPU real: {real_cpu:.2f}% | Prediksi rata-rata: {avg_pred:.2f}%")

    if desired_pods > current_pods:
        print(f"🚀 SCALE UP {current_pods} → {desired_pods}")
        setreplicas(desired_pods)
    elif desired_pods < current_pods:
        print(f"🧊 SCALE DOWN {current_pods} → {desired_pods}")
        setreplicas(desired_pods)
    else:
        print(f"⚖️ CPU stabil, tetap di {current_pods} pods.")

# ==========================================================
# Baca CSV & Simulasi Loop
# ==========================================================
df_all = pd.read_csv(CSV_PATH)
df_all["date"] = pd.to_datetime(df_all["date"])
df_all = df_all.sort_values("date").reset_index(drop=True)

print(f"📁 Data CSV dimuat: {len(df_all)} baris.\n")

for i in range(SEQ_LEN, len(df_all)):
    try:
        # ambil subset data sampai saat ini
        df_cpu_live = df_all.iloc[:i].copy()
        real_cpu_now = df_cpu_live["data"].iloc[-1]

        print("\n==========================================")
        print("⏰", df_cpu_live["date"].iloc[-1], "- update dari CSV")

        # prediksi
        pred_real = predict_cpu(df_cpu_live)

        # tampilkan hasil prediksi sebagian
        last_ts = df_cpu_live["date"].iloc[-1]
        future_timestamps = pd.date_range(start=last_ts, periods=PRED_LEN+1, freq="10min")[1:]
        df_pred = pd.DataFrame({"timestamp": future_timestamps, "predicted_cpu": pred_real})
        print(df_pred.head(3))
        print("...")

        # autoscale
        autoscale_logic(real_cpu_now, pred_real)

        time.sleep(INTERVAL_LOOP)

    except Exception as e:
        print(f"❌ Error: {e}")
        time.sleep(INTERVAL_LOOP)
```
